```python
import tensorflow.keras
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
#define vocabulary size
voc_size=10000
```

```python
#onehot repr
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## word embedding repr
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
embedded_docs
```

```python
##feature represetation
dim=10
model=Sequential()  # You're creating a Sequential model from Keras â€” a simple stack of layers (one after another).

model.add(Embedding(voc_size,dim,input_length=sent_length))  #adding embedding layer to model

'''compiling the model with:

Optimizer: 'adam' (a popular optimizer for training)

Loss: 'mse' (Mean Squared Error loss, normally for regression tasks)'''
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
